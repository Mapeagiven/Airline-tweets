```python
import numpy as pd
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
Dataframe=pd.read_csv('Tweets.csv')
```

```python
Dataframe['airline'].unique()
```

```python
# Filter the DataFrame for 'Virgin America'
df_virgin = Dataframe[Dataframe['airline'] == 'Virgin America']

# Count occurrences of each unique sentiment
sentiment_counts = df_virgin['airline_sentiment'].value_counts().reset_index()
sentiment_counts.columns = ['airline_sentiment', 'count']
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='airline_sentiment', y='count', data=sentiment_counts,color='blue')
plt.title('Sentiment Distribution for Virgin America')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```
